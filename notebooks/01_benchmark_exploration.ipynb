# 01 — Benchmark Exploration

Explore bias benchmark datasets and understand what each benchmark measures.

**Benchmarks covered:**
- BBQ (Bias Benchmark for QA)
- StereoSet (Stereotype preference)
- CrowS-Pairs (Stereotyped sentence preference)
- BOLD (Open-ended generation bias)
- RealToxicityPrompts (Toxic text generation)

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from datasets import load_dataset
from collections import Counter

pd.set_option('display.max_colwidth', 100)
```

## 1. BBQ Dataset
Tests whether models rely on stereotypes in ambiguous contexts.

```python
bbq = load_dataset('heegyu/bbq', split='test')
print(f'BBQ dataset: {len(bbq)} items')
print(f'Columns: {bbq.column_names}')
print(f'\nCategories: {Counter(bbq["category"]).most_common()}')
bbq.to_pandas().head(3)
```

## 2. StereoSet Dataset
Measures whether a model prefers stereotype-consistent completions.

```python
stereoset = load_dataset('stereoset', 'intersentence', split='validation')
print(f'StereoSet: {len(stereoset)} items')
print(f'Bias types: {Counter(stereoset["bias_type"]).most_common()}')
stereoset.to_pandas().head(3)
```

## 3. CrowS-Pairs Dataset
Pairs of sentences: one more stereotypical, one less.

```python
crows = load_dataset('crows_pairs', split='test')
print(f'CrowS-Pairs: {len(crows)} items')
print(f'\nSample pair:')
print(f'  More stereotypical: {crows[0]["sent_more"]}')
print(f'  Less stereotypical: {crows[0]["sent_less"]}')
crows.to_pandas().head(3)
```
